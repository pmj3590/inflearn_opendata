melt를 이용해 df_first와 df_last 데이터 합치기

```python
#pd.melt를 사용하여 df_f_m 라는 변수 만들기
df_f_m=df_first.melt(id_vars="지역", var_name="기간", value_name="평당분양가격")
df_f_m.head()
```

```python
#df_f_m 칼럼 [지역, 기간, 평당분양가격]을 [지역명, 기간, 평당분양가격]으로 변경
df_f_m.columns=["지역명", "기간", "평당분양가격"]
df_f_m.head(1)
```

기간에서 연도와 월을 분리하기

```python
date="2013년12월"
date
```

```python
#split을 이용해서 "년"을 기준으로 텍스트 분리하기
date.split("년")
```

```python
#리스트의 인덱싱을 사용해서 연도만 가져오기
date.split("년")[0]
```

```python
#리스트 인덱싱ㅇ과 replace를 사용해서 월 제거, [-1] : 리스트의 마지막 것을 인출
date.split("년")[-1].replace("월", "")
```

```python
#parse_year
#연도만 반환하도록하고 반환한 데이터는 int타입으로 만들기
def parse_year(date):
    year=date.split("년")[0]
    year=int(year)
    return year
y=parse_year(date)
print(type(y))
```

```python
parse_year(date)
```

```python
#parse_month 변수에 월만 반환하도록 하고 변환 데이터 타입을 int로 변경
def parse_month(date):
    month=date.split("년")[-1].replace("월", "")
    month=int(month)
    return(month)
yy=parse_month(date)
print(type(yy))
```

```python
parse_month(date)
```

```python
#apply를 활용해서 df_f_m 변수의 데이터프레임에 추출한 연도를 새로운 컬럼으로 담기
df_f_m["연도"]=df_f_m["기간"].apply(parse_year)
df_f_m.head(1)
```

```python
#apply를 활용해서 df_f_m 변수의 데이터프레임에 추출한 월를 새로운 컬럼으로 담기
df_f_m["월"]=df_f_m["기간"].apply(parse_month)
df_f_m.head(1)
```

```python
#df_f_m을 df_last와 병합하기 위해서는 컬럼명이 같아야 함
#sample을 활용해서 데이터를 미리보기
df_last.sample()
```

```python
cols=['지역명', '연도', '월','평당분양가격']
cols
```

```python
#df_last 데이터 안에만 전용면적이라는 컬럼이 있음.
#그래서 df_f_m과 맞춰주기 위해 전용면적의 전체만 사용.
#loc를 사용해 전체에 해당하는 면적만 복사하여 df_l_pre변수를 만들기
df_l_pre= df_last.loc[df_last["전용면적"] == "전체", cols].copy()
df_l_pre.head(1)
```

```python
df_f_m.columns
```

```python
cols=['지역명', '연도', '월','평당분양가격']
```

```python
#df_f_m에서 공통된 컬럼만 가져온 뒤
#copy로 복사해서 df_f_pre 컬럼에 담기
df_f_pre=df_f_m[cols].copy()
df_f_pre.head(1)
```
